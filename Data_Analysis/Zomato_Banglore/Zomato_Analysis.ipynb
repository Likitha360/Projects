# Zomato Data Analysis

### Data Cleaning

```python
# Importing Required Libraries
import pandas as pd
import numpy as np
```

```python
#Reading the data
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df['rate'] = df['rate'].replace(['NEW','-'],np.nan)
```

```python
df['rate'] = df['rate'].str.split('/').str[0]
```

```python
df['rate'] = pd.to_numeric(df['rate'], errors = 'coerce')
df['rate']
```

```python
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].str.replace(',','')
df['approx_cost(for two people)']= pd.to_numeric(df['approx_cost(for two people)'],errors='coerce')
```

```python
df = df.dropna(subset=['rate','cuisines','location'])
df['approx_cost(for two people)'] = df['approx_cost(for two people)'].fillna(df['approx_cost(for two people)'].median())
df['rest_type'] = df['rest_type'].fillna('Unknown')
```

```python
df = df.drop_duplicates(subset=['name','location'])
```

```python
df['cuisines'] = df['cuisines'].str.lower()
df['location'] = df['location'].str.lower()
df['rest_type'] = df['rest_type'].str.lower()
```

```python
df.to_csv('zomato_cleaned.csv', index = False)
print('Cleaned data saved successfully.')
```

### Exploring the Data (EDA)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 = pd.read_csv('zomato_cleaned.csv')
```

##### Goal 1: Top cuisines by average rating and volume

```python
df_cuisines = df['cuisines'].str.split(', ',expand=True).stack().reset_index(level=1, drop=True)
df_cuisines.name = 'cuisine'
df_cuisines = df.drop('cuisines', axis = 1).join(df_cuisines)
```

```python
top_cuisines_rating = df_cuisines.groupby('cuisine')['rate'].mean().sort_values(ascending = False).head(5)
print('\nTop 5 cuisines by rating', top_cuisines_rating)
```

```python
top_cuisines_volume = df_cuisines['cuisine'].value_counts().head(5)
print('\n Top 5 cuisines by volume', top_cuisines_volume)
```

```python
# Visulaze of top cuisines by rating
plt.figure(figsize=(10,6))
sns.barplot(x=top_cuisines_rating.index, y=top_cuisines_rating.values)
plt.xticks(rotation=45)
plt.title('Top 5 Cuisines by Average Rating')
plt.xlabel('Cuisine')
plt.ylabel('Average Rating')
plt.savefig('top_cuisines_rating.png')
plt.show()



```

##### Goal 2: Location and cost vs. rating

```python
# correlation b/w cost and rating

plt.figure(figsize=(8,5))
sns.scatterplot(x='approx_cost(for two people)',y='rate', data=df, alpha=0.5)
plt.title('Cost vs. Rating')
plt.xlabel('Approx Cost for Two (INR)')
plt.ylabel('Rating')
plt.savefig('cost_vs_rating.png')
plt.show()
```

```python
avg_rating_by_location = df.groupby('location')['rate'].mean().sort_values(ascending= False).head(5)
print('\n Top 5 locations by average rating', avg_rating_by_location)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=avg_rating_by_location.index, y=avg_rating_by_location.values)
plt.xticks(rotation=45)
plt.title('Top 5 Locations by Average Rating')
plt.xlabel('Location')
plt.ylabel('Average Rating')
plt.savefig('top_locations.png')
plt.show()
```

##### Goal 3: Online ordering and restaurant type vs. rating

```python
online_order_rating = df.groupby('online_order')['rate'].mean()
print("\nAverage rating by online order availability:\n", online_order_rating)
```

```python
rest_type_rating = df.groupby('rest_type')['rate'].mean().sort_values(ascending=False).head(5)
print("\nTop 5 restaurant types by average rating:\n", rest_type_rating)
```

```python
# Visualize restaurant type
plt.figure(figsize=(10, 6))
sns.barplot(x=rest_type_rating.index, y=rest_type_rating.values)
plt.xticks(rotation=45)
plt.title('Top 5 Restaurant Types by Average Rating')
plt.xlabel('Restaurant Type')
plt.ylabel('Average Rating')
plt.savefig('top_rest_types.png')
plt.show()
```

```python
# Top-rated restaurants in a specific location

top_in_koramangala = df[df['location'] == 'koramangala 5th block'].sort_values('rate', ascending=False)[['name', 'rate', 'cuisines', 'approx_cost(for two people)']].head(5)
print("\nTop 5 restaurants in Koramangala 5th Block:\n", top_in_koramangala)
```

```python
# High-rated (4.5+) and affordable (< 500 INR) restaurants
affordable_high_rated = df[(df['rate'] >= 4.5) & (df['approx_cost(for two people)'] <= 500)]\
    [['name', 'rate', 'approx_cost(for two people)', 'location']].sort_values('rate', ascending=False)
print("\nHigh-rated affordable restaurants:\n", affordable_high_rated.head(5))
```

```python
# Restaurants with specific cuisine (e.g., "italian") and high ratings
italian_high_rated = df[df['cuisines'].str.contains('italian') & (df['rate'] >= 4.5)]\
    [['name', 'rate', 'location']].sort_values('rate', ascending=False)
print("\nHigh-rated Italian restaurants:\n", italian_high_rated.head(5))
```
